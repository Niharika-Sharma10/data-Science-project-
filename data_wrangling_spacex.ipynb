```python
# task 4
import pandas as pd

# Step 1: Dataset load karo
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv")

# Step 2: Landing outcomes count karo
landing_outcomes = df['Landing Outcome'].value_counts()
print("Landing Outcomes:\n", landing_outcomes)

# Step 3: Print outcomes with index
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)

# Step 4: Bad outcomes set banao (unsuccessful landings)
bad_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
print("\nBad Outcomes Set:\n", bad_outcomes)

# Step 5: Landing class assign karo (0 = fail, 1 = success)
landing_class = []
for outcome in df['Landing Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

# Step 6: DataFrame me naya column add karo
df['Class'] = landing_class

print("\nFirst 8 rows of Class column:\n", df[['Class']].head(8))
print("\nFirst 5 rows of full DataFrame:\n", df.head(5))

# Step 7: Success rate calculate karo
success_rate = df['Class'].mean()
print("\nSuccess Rate:", success_rate)

# Step 8: CSV export karo
df.to_csv("dataset_part_2.csv", index=False)
print("\nCSV exported as dataset_part_2.csv")
```
